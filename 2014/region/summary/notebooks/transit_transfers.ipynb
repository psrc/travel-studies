```python
def avg_transfers(trips):
    transit_trips = trips[trips['transitsystem1'] > 0 ]
    transfer_transits_2 = transit_trips[transit_trips['transitsystem2'] >0]
    transfer_transits_3 = transit_trips[transit_trips['transitsystem3'] >0]
    transfer_transits_4 = transit_trips[transit_trips['transitsystem4'] >0]

    # total transit trips
    total_trips = transit_trips['expwt_final'].sum()

    # 
    total_transfer_trips_2 = transfer_transits_2['expwt_final'].sum()
    total_transfer_trips_3 = transfer_transits_3['expwt_final'].sum()
    total_transfer_trips_4 = transfer_transits_4['expwt_final'].sum()

    no_transfer = total_trips-total_transfer_trips_2+total_transfer_trips_3+total_transfer_trips_4

    # Average transfers per trip
    return (no_transfer*1+total_transfer_trips_2*2+total_transfer_trips_3*3+total_transfer_trips_4*4)/total_trips
```

```python
avg_transfers(trips)
```
