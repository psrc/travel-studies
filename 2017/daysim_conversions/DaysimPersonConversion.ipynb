```python
import pandas as pd

person = pd.read_csv(r'J:\Projects\Surveys\HHTravel\Survey2017\Data\Dataset_2 August 2017\Person\MattTestPerson.csv')
#household = pd.read_csv(r'J:\Projects\Surveys\HHTravel\Survey2017\Data\Dataset_2 August 2017\Household\MattTestHousehold.csv')
```

```python
final_person = pd.DataFrame()

#hhno
final_person['hhno'] = person['hhid']

#pno
final_person['pno'] = person['pernum']

#pptyp (1 = FTW, 2 = PTW, 3 = non-working, 65+, 4 = non-working, <65, 5 = college student, 6 = HS student, 7 = 5-15, 8 = <5)
#adding columns for categories 6-8 separately based on estimates used for household data
person['hs'] = person['school_loc_name'].str.contains('High School', na = False)

def highschool (row):
   if (row['schooltype'] == 3 or row['schooltype'] == 4 or row['schooltype'] == 5) and row['age'] == 4 :
      return 1
   if (row['schooltype'] == 3 or row['schooltype'] == 4 or row['schooltype'] == 5) and row['age'] == 5 :
      return 1 
   if row['age'] == 3 and row['hs'] is True :
      return 1 
   return 0

person['highschool'] = person.apply (lambda row: highschool (row),axis=1)

def five_to_fifteen (row):
   if row['age'] == 2 or row['age'] == 3  :
      return 1
   return 0

person['five_to_fifteen'] = person.apply (lambda row: five_to_fifteen (row),axis=1)

def under_five (row):
   if row['age'] == 1 :
      return 1
   return 0

person['under_five'] = person.apply (lambda row: under_five (row),axis=1)


def pptyp (row):
    if row['employment'] == 1 :
        return 1
    if row['employment'] == 2 :
        return 2
    if (row['age'] == 10 or row['age'] == 11 or row['age'] == 12) and row['employment'] in range(4,8) :
        return 3 
    if row['age'] in range(5,10) and row['employment'] in range(3,8) :
        return 4 
    if row['schooltype'] == 6 or row['schooltype'] == 7:
        return 5
    if row['highschool'] == 1:
        return 6
    if row['five_to_fifteen'] == 1:
        return 7
    if row['under_five'] == 1:
        return 8
    return -1
    
final_person['pptyp'] = person.apply (lambda row: pptyp (row),axis=1)

#pagey (actual age not available so this is categorical, only other option is to put in average age for each interval)
final_person['pagey'] = person['age']

#pgend (other/no answer = 9 per Daysim guide)
def pgend (row):
    if row['gender'] == 1 :
        return 1
    if row['gender'] == 2 :
        return 2
    return 9

final_person['pgend'] = person.apply (lambda row: pgend (row),axis=1)

#pwtyp
def pwtyp (row):
    if row['employment'] == 1 :
        return 1
    if row['employment'] == 2 :
        return 2
    return 0

final_person['pwtyp'] = person.apply (lambda row: pwtyp (row),axis=1)

#pwpcl (this is not included in the survey data, use Brice's method of linking in correct data)

final_person['pwpcl'] = -1

#pwtaz (this is not included in the survey data, use Brice's method of linking in correct data)

final_person['pwtaz'] = -1

#pwautime (not sure how to calculate)

final_person['pwautime'] = -1

#pwaudist (not sure how to calculate)

final_person['pwaudist'] = -1

#pstyp
def pstyp (row):
    if row['schooltype'] == 3 :
        return 1
    if row['employment'] == 2 :
        return 2
    return 0

final_person['pstyp'] = person.apply (lambda row: pstyp (row),axis=1)

#pspcl (this is not included in the survey data, use Brice's method of linking in correct data)

final_person['pspcl'] = -1

#pstaz (this is not included in the survey data, use Brice's method of linking in correct data)

final_person['pstaz'] = -1

#psautime (not sure how to calculate)

final_person['psautime'] = -1

#psaudist (not sure how to calculate)

final_person['psaudist'] = -1

#puwmode (the Daysim guidelines do not match up with survey data, alternative groupings are used here)
# 1 = walk, 2 = bike, 3 = drive alone, 4 = carpool, 5 = transit, 0 = other

def puwmode (row):
    if row['commute_mode'] == 7 :
        return 1
    if row['commute_mode'] == 6 :
        return 4
    if row['commute_mode'] == 1 or row['commute_mode'] == 4 :
        return 3
    if row['commute_mode'] == 2 or row['commute_mode'] == 3 or row['commute_mode'] == 5 :
        return 4
    if row['commute_mode'] in range(8,15) :
        return 5
    return 0
    
final_person['puwmode'] = person.apply (lambda row: puwmode (row),axis=1)

#puwarrp (this field does not seem to be included in this year's survey)

final_person['puwarrp'] = -1

#puwdepp (this field does not seem to be included in this year's survey)

final_person['puwdepp'] = -1

#ptpass (I take this to mean that the person is USING a transit pass provided/subsidized by employer/school)
def ptpass (row):
    if row['tran_pass_12'] == 1 or row['tran_pass_12'] == 2 or row['benefits_3'] == 3 :
        return 1
    return 0

final_person['ptpass'] = person.apply (lambda row: ptpass (row),axis=1)

#ppaidpark (I take this to mean that there is paid parking at work, regardless of who is paying for it)
def ppaidpark (row):
    if row['workpass'] in range(1,5) :
        return 1
    return 0

final_person['ppaidpark'] = person.apply (lambda row: ppaidpark (row),axis=1)

#pdiary (this field does not seem to be included in this year's survey, also probably not relevant)

final_person['pdiary'] = -1

#pproxy
def pproxy (row):
    if row['proxy'] == 1 :
        return 1
    if row['proxy'] == 2 or row['proxy'] == 3 :
        return 2
    return -1

final_person['pproxy'] = person.apply (lambda row: pproxy (row),axis=1)

#psexpfac (this field does not seem to be included in this year's survey)

final_person['psexpfac'] = -1

```
